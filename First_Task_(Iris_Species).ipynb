Importing Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the Uplaoded File, i.e, Iris.csv

```python
iris = pd.read_csv('Iris.csv')
```

**Displaying Dataset Structure**

Using head()

```python
print(iris.head())
```

using shape

```python
print("Shape: ", iris.shape)
```

Using Columns

```python
print("Columns: ",iris.columns)
```

Creating Scatter plot

```python
plt.figure(figsize = (8,5))
sns.scatterplot(data = iris, x = "SepalLengthCm", y = "SepalWidthCm", hue = "Species")
plt.title("Sepal Length Vs Sepal Width by species")
sns.pairplot(iris, hue="Species")
plt.show()
plt.show
```

Creating Histogram

```python
plt.figure(figsize = (8,5))
sns.histplot(iris['PetalLengthCm'], bins = 20, kde = True)
plt.title("Petal Length Histogram")
plt.show()
```

Creating BoxPlot

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = "Species", y = "PetalWidthCm", data = iris)
plt.title("Boxplot of Petal Width Species")
plt.show()
```
